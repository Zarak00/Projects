```python
customer_to_remove = ["VistaJet Luftfahrtunternehmen GmbH", "XOJET Aviation LLC", "Western Air Charter Inc dba Jet Edge"]
df1=df1[~df1["CUSTOMER"].isin(customer_to_remove)]
```

```python
MRKT_NM_to_remove = ['ADC DITCO', 'ADC DITCO ']
df1=df1[~df1["MRKT_NM"].isin(MRKT_NM_to_remove)]
```

```python
df1["MRKT_NM"].unique()
```

```python
df1['ROYALTY ELIGIBLE AMT'] = np.where(df1['ELIGIBLE'] == 'Y', df1['AMT'] * df1['APPLICABLE_PERCENTAGE'], 0)
```

```python
df1['ROYALTY AMT'] = df1['ROYALTY ELIGIBLE AMT'] * 0.08
```

# Values to look up for in Tail Eligable

```python
df1["SERVICE"] = df1["SERVICE"].str.upper()
df2["SERVICE"] = df2["SERVICE"].str.upper()
```

```python
df2['SERVICE'] = df2['SERVICE'].replace('VIASAST', 'VIASAT')
```

```python
df2["SERVICE"].unique()
```

```python
def vlookup(x):
    unique_services = df2["SERVICE"].unique()
    if x["SERVICE"] in unique_services:
        try:
            return df2.loc[df2["Tail ID"] == x["ADC_TAIL_ID"], "EXCLUDE"].iloc[0]
        except IndexError:
            return "Y"
    else:
        return "Y"

# Applying the VLOOKUP equivalent function to df1
df1["TAIL ELIGIBLE"] = df1.apply(vlookup, axis=1)
```

```python
filtered_df = df1[df1['TAIL ELIGIBLE'] == 'N']
filtered_df.head(15)
```

```python
df1.to_excel('Bombardier_report.xlsx', index=False, sheet_name='DATA', engine='openpyxl')

# Load the Excel file using openpyxl to auto-adjust column widths
from openpyxl import load_workbook

# Load the workbook
workbook = load_workbook('Bombardier_report.xlsx')
worksheet = workbook.active

# Auto-adjust the column width for all columns
for column_cells in worksheet.columns:
    length = max(len(str(cell.value)) for cell in column_cells)
    worksheet.column_dimensions[column_cells[0].column_letter].width = length + 2

# Save the workbook back to the Excel file
workbook.save('Bombardier_report.xlsx')
print("Bombardier report has been exported to Excel successfully!")
```

# Creating Inovices

```python
uploaded = files.upload()
```

```python

# Assuming you have uploaded the file and assigned its name to 'file_name'
file_name = list(uploaded.keys())[0]

# Specify the data types for the columns you want to preserve leading zeros
dtype_dict = {'PROJECT CODE': str}

# Read the Excel file with the specified data types
df3 = pd.read_excel(file_name, dtype=dtype_dict)

df3.head(1)
```

```python
df3.tail(16)
```

```python
df4 = df3.groupby(["PROJECT CODE"])["TOTAL ROYALTY"].sum().reset_index()
df4
```

```python
total_royalty_sum = df4['TOTAL ROYALTY'].sum()
df4.loc[len(df4)] = ['Total', total_royalty_sum]
df4
```

```python
def map_project_name(project_number):
    PROJECT_NAME = {
        '000358': "VHF Datalink",
        '000359': "SAT Datalink",
        '000264': "JX",
        '000292': "Lux",
        '000322': "Ka"
    }

    return PROJECT_NAME.get(str(project_number), "Unknown")

df4["PROJECT_NAME"] = df4["PROJECT CODE"].map(map_project_name)
```

```python
df4
```

```python
writer = pd.ExcelWriter('FINAL_Bombardier_report.xlsx', engine='openpyxl')
df1.to_excel(writer, index=False, sheet_name='DATA')
df4.to_excel(writer, index=False, sheet_name='INVOICE')

# Close the Pandas Excel writer
writer.close()

# Load the workbook
workbook = load_workbook('FINAL_Bombardier_report.xlsx')

# Auto-adjust the column width for all sheets
for sheet in workbook.sheetnames:
    worksheet = workbook[sheet]
    for column_cells in worksheet.columns:
        length = max(len(str(cell.value)) for cell in column_cells)
        worksheet.column_dimensions[column_cells[0].column_letter].width = length + 2

# Save the workbook back to the Excel file
workbook.save('FINAL_Bombardier_report.xlsx')
print("FINAL_Bombardier report has been exported to Excel successfully!")
```
