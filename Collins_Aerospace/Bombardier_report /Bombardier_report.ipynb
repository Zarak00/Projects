```python
df1 = pd.read_excel("/content/Bombardier - July 2023 Data.xlsx", sheet_name = "Sheet1")
df1.head(2)
```

```python
df2 = pd.read_excel("/content/TAIL ELIGIBLE.xlsx")
df2.head(3)
```

```python
customer_to_remove = ["VistaJet Luftfahrtunternehmen GmbH", "XOJET Aviation LLC", "Western Air Charter Inc dba Jet Edge"]
df1=df1[~df1["CUSTOMER"].isin(customer_to_remove)]
```

```python
MRKT_NM_to_remove = ['ADC DITCO', 'ADC DITCO ']
df1=df1[~df1["MRKT_NM"].isin(MRKT_NM_to_remove)]
```

```python
df1["MRKT_NM"].unique()
```

```python
df1['ROYALTY ELIGIBLE AMT'] = np.where(df1['ELIGIBLE'] == 'Y', df1['AMT'] * df1['APPLICABLE_PERCENTAGE'], 0)
```

```python
df1['ROYALTY AMT'] = df1['ROYALTY ELIGIBLE AMT'] * 0.08
```

```python
df1.head(5)
```

```python
df1.info()
```

```python
def vlookup(x):
    if x["SERVICE"] in ["JX", "Viasat", "Lux"]:
        try:
            return df2.loc[df2["Tail ID"] == x["ADC_TAIL_ID"], "EXCLUDE"].iloc[0]
        except IndexError:
            return "Y"
    else:
        return "Y"

# Applying the VLOOKUP equivalent function to df1
df1["TAIL ELIGIBLE"] = df1.apply(vlookup, axis=1)
```
