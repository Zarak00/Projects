# ML

```python
# Import necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Extract features (X) and target variables (y)
X = tetouan_monday_resampled[['Temperature', 'Humidity', 'Wind Speed', 'general diffuse flows', 'diffuse flows']]
y_zone1 = tetouan_monday_resampled['Zone 1 Power Consumption']
y_zone2 = tetouan_monday_resampled['Zone 2  Power Consumption']
y_zone3 = tetouan_monday_resampled['Zone 3  Power Consumption']

# Split the data into training and testing sets (80% training, 20% testing)
X_train, X_test, y_zone1_train, y_zone1_test, y_zone2_train, y_zone2_test, y_zone3_train, y_zone3_test = train_test_split(
    X, y_zone1, y_zone2, y_zone3, test_size=0.2, random_state=42
)

# Choose a machine learning model (Linear Regression in this case)
model_zone1 = LinearRegression()
model_zone2 = LinearRegression()
model_zone3 = LinearRegression()

# Train the model on the training data
model_zone1.fit(X_train, y_zone1_train)
model_zone2.fit(X_train, y_zone2_train)
model_zone3.fit(X_train, y_zone3_train)

# Make predictions on the testing data
y_zone1_pred = model_zone1.predict(X_test)
y_zone2_pred = model_zone2.predict(X_test)
y_zone3_pred = model_zone3.predict(X_test)

# Evaluate the model using Mean Squared Error (MSE)
mse_zone1 = mean_squared_error(y_zone1_test, y_zone1_pred)
mse_zone2 = mean_squared_error(y_zone2_test, y_zone2_pred)
mse_zone3 = mean_squared_error(y_zone3_test, y_zone3_pred)

print(f'Mean Squared Error (Zone 1): {mse_zone1}')
print(f'Mean Squared Error (Zone 2): {mse_zone2}')
print(f'Mean Squared Error (Zone 3): {mse_zone3}')
```

# Accuracy

```python
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Create bins for power consumption levels (you can adjust these bins based on your specific criteria)
bins = [0, 100, 200, 300, float('inf')]
labels = ['Very Low', 'Low', 'Medium', 'High']

# Convert predicted power consumption values to categories for each zone
y_zone1_pred_category = pd.cut(y_zone1_pred, bins=bins, labels=labels)
y_zone2_pred_category = pd.cut(y_zone2_pred, bins=bins, labels=labels)
y_zone3_pred_category = pd.cut(y_zone3_pred, bins=bins, labels=labels)

# Convert power consumption to categories for the test data (if you haven't done it already)
y_zone1_test_category = pd.cut(y_zone1_test, bins=bins, labels=labels)
y_zone2_test_category = pd.cut(y_zone2_test, bins=bins, labels=labels)
y_zone3_test_category = pd.cut(y_zone3_test, bins=bins, labels=labels)

# Calculate accuracy scores for the test data
accuracy_zone1 = accuracy_score(y_zone1_test_category, y_zone1_pred_category)
accuracy_zone2 = accuracy_score(y_zone2_test_category, y_zone2_pred_category)
accuracy_zone3 = accuracy_score(y_zone3_test_category, y_zone3_pred_category)

print(f'Accuracy Score (Zone 1): {accuracy_zone1:.2f}')
print(f'Accuracy Score (Zone 2): {accuracy_zone2:.2f}')
print(f'Accuracy Score (Zone 3): {accuracy_zone3:.2f}')
# Visualization of Predictions for the test data
plt.figure(figsize=(10, 8))

# Plot actual vs predicted for Zone 1
plt.subplot(3, 1, 1)
plt.scatter(y_zone1_test, y_zone1_pred, alpha=0.5)
plt.plot([min(y_zone1_test), max(y_zone1_test)], [min(y_zone1_test), max(y_zone1_test)], '--k')
plt.title='Zone 1: Actual vs Predicted'
plt.xlabel='Actual Power Consumption'
plt.ylabel='Predicted Power Consumption'

# Plot actual vs predicted for Zone 2
plt.subplot(3, 1, 2)
plt.scatter(y_zone2_test, y_zone2_pred, alpha=0.5)
plt.plot([min(y_zone2_test), max(y_zone2_test)], [min(y_zone2_test), max(y_zone2_test)], '--k')
plt.title='Zone 2: Actual vs Predicted'
plt.xlabel='Actual Power Consumption'
plt.ylabel='Predicted Power Consumption'

# Plot actual vs predicted for Zone 3
plt.subplot(3, 1, 3)
plt.scatter(y_zone3_test, y_zone3_pred, alpha=0.5)
plt.plot([min(y_zone3_test), max(y_zone3_test)], [min(y_zone3_test), max(y_zone3_test)], '--k')
plt.title='Zone 3: Actual vs Predicted'
plt.xlabel='Actual Power Consumption'
plt.ylabel='Predicted Power Consumption'

plt.tight_layout()
plt.show()
```
